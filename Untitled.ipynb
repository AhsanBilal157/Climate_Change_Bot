```python
import os
import requests
import re

# Define the folder to store downloaded PDFs
output_folder = 'downloaded_pdfs'
os.makedirs(output_folder, exist_ok=True)

# Define a text file to store URLs that fail to download
failed_log_file = os.path.join(output_folder, 'failed_downloads.txt')

# Define a function to sanitize filenames by replacing invalid characters
def sanitize_filename(filename):
    # Remove characters that are invalid in filenames
    return re.sub(r'[<>:"/\\|?*]', '_', filename)

# Define a function to download a PDF given its URL
def download_pdf(url, folder):
    # Sanitize and extract the filename
    filename = os.path.join(folder, sanitize_filename(url.split("/")[-1]))
    try:
        # Send a GET request to download the PDF
        response = requests.get(url)
        response.raise_for_status()  # Check if the request was successful
        
        # Save the PDF content to a file
        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {filename}")
    except requests.exceptions.RequestException as e:
        # Log failed downloads
        print(f"Failed to download {url}: {e}")
        with open(failed_log_file, 'a') as log_file:
            log_file.write(f"{url}\n")

# Extract unique links and remove NaN values
pdf_links = df['document_metadata.source_url'].dropna().unique()

# Download each unique PDF link
for link in pdf_links:
    download_pdf(link, output_folder)

print(f"Downloads completed. Check '{failed_log_file}' for any failed downloads.")
```
